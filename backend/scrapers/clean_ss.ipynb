```python
import json
import re

with open("ss.json") as f:
    data = json.load(f)
```

```python
res = []
for category, category_data in data.items():
    for item in category_data:
        product = {}
        product["barcodes"] = None
        product["market_id"] = 3
        product["brand"] = item.get("brand")
        product["price"] = item.get("price")
        product["availability"] = not item.get("isSoldOut")
        product["name"] = item.get("name")
        assert "imgKey" in item
        product["image"] = f"https://ssecomm.s3-ap-southeast-1.amazonaws.com/products/md/{item.get('imgKey')}.0.jpg"
        product["quantity"] = item.get("packSize")
        product["product_url"] = "https://shengsiong.com.sg/product/" + item.get("slug")
        if item["associatedPromoRules"]:
            for k, v in item["associatedPromoRules"].items():
                label = v['label']
                print(label)
                break
            find = re.findall("Buy (\d+) for \$(\d+\.\d*)", label)
            if find:
                find = find[0]
                product['offer_qty'] = int(find[0])
                product['offer_price'] = float(find[1])
                product['offer_desc'] = label
        else:
            product['offer_qty'] = None
            product['offer_price'] = None
            product['offer_desc'] = None
        res.append(product)

            
```

```python
with open("clean_ss.json", "w") as f:
    json.dump(res, f)
```
